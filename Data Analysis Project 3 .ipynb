### Question ( mapping + data-type casting )
# 4. What is the mean stop_duration ?


```python
# df['Column_name'] = df['Column_name'].map( { old:new, old:new})

# df['Column_name'].mean()
```

```python
data.head()
```

```python
# need to convert column stop duration to integer to run mean 
data.stop_duration.unique()
```

```python
data.stop_duration.value_counts()
```

```python
data
```

```python
data['stop_duration'].mean()  # mean of stop duratioin is 12.187 
```
